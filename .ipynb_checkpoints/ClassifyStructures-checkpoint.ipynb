```python
import pyAgrum as gum
import pyAgrum.lib.notebook as gnb
import pyAgrum.lib.bn_vs_bn as bnvsbn

import os
import pandas as pd
import numpy as np
```

```python
structure_files = os.listdir("outputs/structures")
len(structure_files)
```

```python
#TODO SPLIT ON INPUT PARAMETERS (THAT ARE OBSERVABLE FROM A DATASET)
structures = {}
scores = os.listdir("outputs/scores")
for f in structure_files:
    structure, rep = f[9:-7].split("_")
    if f"structure{structure}_{rep}.csv" in scores:
        n, t, v, a, s = pd.read_csv(f"outputs/scores/structure{structure}_{rep}.csv")[["nodes", "timesteps", "values", "algorithm", "score"]].iloc[0].values
        if f"{n}.{t}.{v}.{a}.{s}" in structures:
            structures[f"{n}.{t}.{v}.{a}.{s}"][0].append(structure)
            structures[f"{n}.{t}.{v}.{a}.{s}"][1].append(rep)
        else:
            structures[f"{n}.{t}.{v}.{a}.{s}"] = [[structure],[rep]]
#         print(n, t, v, a, s)
#         break
structures
```

```python
dfs = {}
for parameters in structures:
#     print(structures[parameters])
    df = pd.DataFrame(np.transpose(structures[parameters]), columns=["structure","rep"])
    dfs[parameters] = df
#     break
dfs[parameters]
    
```

```python
def read_score(row):
    return np.mean(pd.read_csv(f"outputs/scores/structure{row['structure']}_{row['rep']}.csv", index_col=0)["fscore"])
#TODO: MEAN PAKKEN OF DIE VAN GROOTSTE AANTAL ITEMS? 
```

```python
for parameters in dfs:
    dfs[parameters]["fscore"] = dfs[parameters].apply(read_score, axis=1)
    dfs[parameters]
dfs
```

```python
def eigenvalues(row):
    bn = gum.BayesNet()
    bn.loadBIFXML(f"outputs/structures/structure{row['structure']}_{row['rep']}.bifxml")
    
    data,_ = gum.generateSample(bn,1000,None,False)
    data = data.reindex(sorted(data.columns, key=lambda x: (len(x), int(x.split(".")[1]), x.split(".")[0])), axis=1)
    data = data.astype('int')
    
    cov = data.cov()
    
    eigenvalues = np.linalg.eig(cov)[0]
    return pd.Series([np.mean(eigenvalues),np.var(eigenvalues)])#sorted(eigenvalues, reverse=True))

# df2 = pd.concat([df, df.apply(eigenvalues, axis=1)], axis=1)
# display(df2.style.background_gradient(cmap='coolwarm'))


def nr_eigenvalues(row):
    bn = gum.BayesNet()
    bn.loadBIFXML(f"outputs/structures/structure{row['structure']}_{row['rep']}.bifxml")
    
    data,_ = gum.generateSample(bn,1000,None,False)
    data = data.reindex(sorted(data.columns, key=lambda x: (len(x), int(x.split(".")[1]), x.split(".")[0])), axis=1)
    data = data.astype('int')
    
    cov = data.cov()
    
    eigenvalues = np.linalg.eig(cov)[0]
    return len(eigenvalues)

# df3 = pd.concat([df, df.apply(nr_eigenvalues, axis=1)], axis=1)


def minmax_eigenvalues(row):
    bn = gum.BayesNet()
    bn.loadBIFXML(f"outputs/structures/structure{row['structure']}_{row['rep']}.bifxml")
    
    data,_ = gum.generateSample(bn,1000,None,False)
    data = data.reindex(sorted(data.columns, key=lambda x: (len(x), int(x.split(".")[1]), x.split(".")[0])), axis=1)
    data = data.astype('int')
    
    cov = data.cov()
    
    eigenvalues = np.linalg.eig(cov)[0]
    return pd.Series([min(eigenvalues),max(eigenvalues)])

# df4 = pd.concat([df, df.apply(minmax_eigenvalues, axis=1)], axis=1)
```

```python
eigenvalues_dfs = {}

for parameters in dfs:

    eigenvalues_dfs[parameters] = pd.concat([dfs[parameters], dfs[parameters].apply(eigenvalues, axis=1)], axis=1)
    print(parameters)
    display(eigenvalues_dfs[parameters].corr()[["fscore"]][-2:].style.background_gradient(cmap='coolwarm', vmin=-1, vmax=1))
# df4[["fscore",0,1]].corr()[["fscore"]].style.background_gradient(cmap='coolwarm')
```

```python
nr_eigenvalues_dfs = {}

for parameters in dfs:

    nr_eigenvalues_dfs[parameters] = pd.concat([dfs[parameters], dfs[parameters].apply(nr_eigenvalues, axis=1)], axis=1)
    print(parameters)
    display(nr_eigenvalues_dfs[parameters].corr()[["fscore"]][-1:].style.background_gradient(cmap='coolwarm', vmin=-1, vmax=1))
```

```python
minmax_eigenvalues_dfs = {}

for parameters in dfs:

    minmax_eigenvalues_dfs[parameters] = pd.concat([dfs[parameters], dfs[parameters].apply(minmax_eigenvalues, axis=1)], axis=1)
    print(parameters)
    display(minmax_eigenvalues_dfs[parameters].corr()[["fscore"]][-2:].style.background_gradient(cmap='coolwarm', vmin=-1, vmax=1))
```
