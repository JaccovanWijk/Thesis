```python
all_algorithms = df["algorithm"].unique()
all_scorefunctions = df["scorefunction"].unique()

plt.figure()

colors = ['blue','red','orange', 'green', 'cyan']
for a, algorithm in enumerate(all_algorithms):
    for s, scorefunction in enumerate(all_scorefunctions):
        sub = df[((df["algorithm"]==algorithm)&(df["scorefunction"]==scorefunction))]
        print((a - len(all_algorithms)/2)*300 + 150)
        if algorithm == all_algorithms[0]:
            plt.bar(sub[["algorithm","fscore"]].groupby("algorithm").mean().index,# + (a - len(all_algorithms)/2)*300 + 150, 
                    sub[["algorithm","fscore"]].groupby("algorithm").mean()["fscore"].values, 
                    yerr=sub[["algorithm","fscore"]].groupby("algorithm").var()["fscore"].values,
                    align='center',
                    alpha=0.5,
                    ecolor='black',
                    width=300,
                    color=colors[a],
                    label=f"{scorefunction}")
        else:
            plt.bar(sub[["algorithm","fscore"]].groupby("algorithm").mean().index,# + (a - len(all_algorithms)/2)*300 + 150, 
                    sub[["algorithm","fscore"]].groupby("algorithm").mean()["fscore"].values, 
                    yerr=sub[["algorithm","fscore"]].groupby("algorithm").var()["fscore"].values,
                    align='center',
                    alpha=0.5,
                    ecolor='black',
                    width=300,
                    color=colors[a])
        
plt.xlabel("items")
plt.ylabel("fscore")
plt.legend(bbox_to_anchor=[1,1])
plt.show()
```

```python
all_items = df["items"].unique()
all_nodes = df["nodes"].unique()

plt.figure()

colors = ['blue','red','orange']
for n, nodes in enumerate(all_nodes):
    for i, items in enumerate(all_items):
        sub = df[((df["items"]==items)&(df["nodes"]==nodes))]
        
        if items == all_items[0]:
            plt.bar(sub[["items","fscore"]].groupby("items").mean().index + (n - len(all_nodes)/2)*300 + 150, 
                    sub[["items","fscore"]].groupby("items").mean()["fscore"].values, 
                    yerr=sub[["items","fscore"]].groupby("items").var()["fscore"].values,
                    align='center',
                    alpha=0.5,
                    ecolor='black',
                    width=300,
                    color=colors[n],
                    label=f"{nodes} nodes")
        else:
            plt.bar(sub[["items","fscore"]].groupby("items").mean().index + (n - len(all_nodes)/2)*300 + 150, 
                    sub[["items","fscore"]].groupby("items").mean()["fscore"].values, 
                    yerr=sub[["items","fscore"]].groupby("items").var()["fscore"].values,
                    align='center',
                    alpha=0.5,
                    ecolor='black',
                    width=300,
                    color=colors[n])
        
plt.xlabel("items")
plt.ylabel("fscore")
plt.legend(bbox_to_anchor=[1,1])
plt.show()


plt.figure()
all_times = df["timesteps"].unique()

colors = ['blue','red','orange']
for t, times in enumerate(all_times):
    for i, items in enumerate(all_items):
        sub = df[((df["items"]==items)&(df["timesteps"]==times))]
        
        if items == all_items[0]:
            plt.bar(sub[["items","fscore"]].groupby("items").mean().index + (t - len(all_times)/2)*300 + 150, 
                    sub[["items","fscore"]].groupby("items").mean()["fscore"].values, 
                    yerr=sub[["items","fscore"]].groupby("items").var()["fscore"].values,
                    align='center',
                    alpha=0.5,
                    ecolor='black',
                    width=300,
                    color=colors[t],
                    label=f"{times} timesteps")
        else:
            plt.bar(sub[["items","fscore"]].groupby("items").mean().index + (t - len(all_times)/2)*300 + 150, 
                    sub[["items","fscore"]].groupby("items").mean()["fscore"].values, 
                    yerr=sub[["items","fscore"]].groupby("items").var()["fscore"].values,
                    align='center',
                    alpha=0.5,
                    ecolor='black',
                    width=300,
                    color=colors[t])
        
plt.xlabel("items")
plt.ylabel("fscore")
plt.legend(bbox_to_anchor=[1,1])
plt.show()



plt.figure()
all_values = df["values"].unique()

colors = ['blue','red','orange']
for v, values in enumerate(all_values):
    for i, items in enumerate(all_items):
        sub = df[((df["items"]==items)&(df["values"]==values))]
        
        if items == all_items[0]:
            plt.bar(sub[["items","fscore"]].groupby("items").mean().index + (v - len(all_values)/2)*300 + 150, 
                    sub[["items","fscore"]].groupby("items").mean()["fscore"].values, 
                    yerr=sub[["items","fscore"]].groupby("items").var()["fscore"].values,
                    align='center',
                    alpha=0.5,
                    ecolor='black',
                    width=300,
                    color=colors[v],
                    label=f"{values} values")
        else:
            plt.bar(sub[["items","fscore"]].groupby("items").mean().index + (v - len(all_values)/2)*300 + 150, 
                    sub[["items","fscore"]].groupby("items").mean()["fscore"].values, 
                    yerr=sub[["items","fscore"]].groupby("items").var()["fscore"].values,
                    align='center',
                    alpha=0.5,
                    ecolor='black',
                    width=300,
                    color=colors[v])
        
plt.xlabel("items")
plt.ylabel("fscore")
plt.legend(bbox_to_anchor=[1,1])
plt.show()




plt.figure()
algorithms = df["algorithm"].unique()

colors = ['blue','red','orange', 'green', 'cyan']
for a, algorithm in enumerate(algorithms):
    for i, items in enumerate(all_items):
        sub = df[((df["items"]==items)&(df["algorithm"]==algorithm))]
        
        if items == all_items[0]:
            plt.bar(sub[["items","fscore"]].groupby("items").mean().index + (a - len(algorithms)/2)*150 + 75, 
                    sub[["items","fscore"]].groupby("items").mean()["fscore"].values, 
                    yerr=sub[["items","fscore"]].groupby("items").var()["fscore"].values,
                    align='center',
                    alpha=0.5,
                    ecolor='black',
                    width=150,
                    color=colors[a],
                    label=f"{algorithm} algorithm")
        else:
            plt.bar(sub[["items","fscore"]].groupby("items").mean().index + (a - len(algorithms)/2)*150 + 75, 
                    sub[["items","fscore"]].groupby("items").mean()["fscore"].values, 
                    yerr=sub[["items","fscore"]].groupby("items").var()["fscore"].values,
                    align='center',
                    alpha=0.5,
                    ecolor='black',
                    width=150,
                    color=colors[a])
        
plt.xlabel("items")
plt.ylabel("fscore")
plt.legend(bbox_to_anchor=[1,1])
plt.show()


scores = df["score"].unique()

colors = ['blue','red','orange', 'green', 'cyan']
for s, score in enumerate(scores):
    for i, items in enumerate(all_items):
        sub = df[((df["items"]==items)&(df["score"]==score))]
        
        if items == all_items[0]:
            plt.bar(sub[["items","fscore"]].groupby("items").mean().index + (s - len(scores)/2)*150 + 75, 
                    sub[["items","fscore"]].groupby("items").mean()["fscore"].values, 
                    yerr=sub[["items","fscore"]].groupby("items").var()["fscore"].values,
                    align='center',
                    alpha=0.5,
                    ecolor='black',
                    width=150,
                    color=colors[s],
                    label=f"{score} metric")
        else:
            plt.bar(sub[["items","fscore"]].groupby("items").mean().index + (s - len(scores)/2)*150 + 75, 
                    sub[["items","fscore"]].groupby("items").mean()["fscore"].values, 
                    yerr=sub[["items","fscore"]].groupby("items").var()["fscore"].values,
                    align='center',
                    alpha=0.5,
                    ecolor='black',
                    width=150,
                    color=colors[s])
        
plt.xlabel("items")
plt.ylabel("fscore")
plt.legend(bbox_to_anchor=[1,1])
plt.show()
```
