```python
import pandas as pd
import os
```

```python
files = os.listdir("outputs/scores")
len(files)
```

```python
df = pd.read_csv('outputs/scores/'+files[0], index_col=0)
for file in files[1:]:
    df = pd.concat([df, pd.read_csv('outputs/scores/'+file, index_col=0)], ignore_index=True)
df
```
