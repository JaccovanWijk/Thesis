```python
%%time
from syntheticDBN2 import Bayes_Test
import pandas as pd
import matplotlib.pyplot as plt
import pyAgrum.lib.notebook as gnb
%matplotlib widget
```

```python
%%time
# Create True DBN
timesteps = 3
nodes = 4
folds = 10
bayestest = Bayes_Test(timesteps=timesteps, nodes=nodes, structure='serie', values=7, folds=folds) 

gnb.show(bayestest.true_dbn)
```

```python
bayestest.generate_data(train_items=100000)
```

```python
arcs = {}
total = 0
for t in range(timesteps):
    for f in range(folds):
        bn = bayestest.learn_causal_structure(t=t, fold=f)
        for arc in bn.arcs():
            if arc in arcs:
                arcs[arc] += 1
            else:
                arcs[arc] = 1
#         gnb.show(bn)
#         print(bn.arcs())
        total += 1
    
history = []
for arc1 in arcs:
    for arc2 in arcs:
        if arc1[0] == arc2[1] and arc1[1] == arc2[0] and arc1 not in history:
            print(arc1, round(arcs[arc1]/total*100), arc2, round(arcs[arc2]/total*100))
            history += [arc1, arc2]
```
