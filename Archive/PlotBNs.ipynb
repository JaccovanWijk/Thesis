```python
from causalscore import *
import pyAgrum.lib.notebook as gnb
import pyAgrum.lib.bn_vs_bn as bnvsbn
from pyAgrum.lib import image
import random
```

```python
nodes = 5

# cmap = np.delete(np.arange(0.05,1.05,0.05),[10,11,12])
# names = [f"{chr(ord('`')+(x%26 + 1))}"*(np.ceil((x+1)/26).astype(int)) for x in range(nodes)]
# # colors = {name:(i+1)/(len(names)*2+1) for i, name in enumerate(names)}
# colors = {name:random.choice(cmap) for name in names}
# colors
colors = {'a': 0.1, 'b': 0.3, 'c': 0.5, 'd': 0.9500000000000001, 'e': 0.2}
```

```python

generator = Generator(nodes=nodes, timesteps=3, values=2, arcs_ratio=1.1, islands=1, algorithm="TABU", score="BDEU")
twodbn = generator.generate_structure()
true_dbn = generator.generate_DBN(twodbn)
# gnb.show(bn)


all_colors = {}
for name in true_dbn.names():
    all_colors[name] = colors[name.split(".")[0]]
    
print(all_colors)

gnb.show(true_dbn)
image.export(true_dbn,"Figs/true_dbn_example.png", nodeColor=all_colors)
```

```python
data = generator.generate_data(items=100000)
data
```

```python
causal = generator.learn_causal_structure(100000)
gnb.show(causal)
image.export(causal,"Figs/learned_causal_example.png", nodeColor=all_colors)
```

```python
learned = generator.learn_time_structure(100000,times=2)
gnb.show(learned)
image.export(learned,"Figs/learned_time_example.png", nodeColor=all_colors)
```

```python
learned = generator.get_structure()
gnb.show(learned)
image.export(learned,"Figs/learned_example.png", nodeColor=all_colors)
```

```python
gcmp=bnvsbn.GraphicalBNComparator(true_dbn,learned)
gnb.flow.add(bnvsbn.graphDiff(true_dbn,learned))
gnb.flow.add(bnvsbn.graphDiffLegend())
gnb.flow.new_line()
gnb.flow.add_html("<br/>".join([f"{k} : {v:.2f}" for k,v in gcmp.skeletonScores().items() if k!='count']),"Skeleton scores")
gnb.flow.add_html("<br/>".join([f"{k} : {v:.2f}" for k,v in gcmp.scores().items() if k!='count']),"Parameter scores")

gnb.flow.display()
```
