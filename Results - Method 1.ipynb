```python
df = pd.read_csv("newOutputs/classifiers.csv", index_col=0)
df
```

# Bin Continuous Properties

```python
bins = 4

for col in ["PCweight1","PCweight2","PCweight3","PCweight4","PCAsAbove95", "maxdegree/nodes","roots/nodes","leaves/nodes","longestpath/nodes","mean_constraint","mean_degree","PHI","items/PHI"]:
    if len(df[col].unique()) > bins:
        combinations = np.linspace(min(df[col]),max(df[col]),bins+1)
        df[col] = pd.cut(df[col], combinations, include_lowest=True, labels=[f"{np.round(combinations[x],2)}-{np.round(combinations[x+1],2)}" for x in range(len(combinations)-1)]) 
df
```

# Plot Property

```python
parameter = "items"
save_name = "items"
label_name = "Datapoints"

colors = {"L2L":"tab:purple", "AIC":"tab:blue", "BIC":"tab:green", "BDEU":"tab:orange", "K2":"tab:red", "MIIC":"tab:brown", "3OFF2":"tab:olive"}

all_algorithms = [["HC"], ["TABU"], ["MMHC"],["MIIC","3OFF2"]]
for i, algs in enumerate(all_algorithms):   
#     display(df[["algorithm","scorefunction","fscore",parameter]].groupby(["algorithm","scorefunction", parameter]).var()[["fscore"]].dropna())
    grouped = df[["algorithm","scorefunction","fscore",parameter]].groupby(["algorithm","scorefunction", parameter]).mean()[["fscore"]]
    maxes = grouped[grouped["fscore"].isin(grouped.groupby(parameter)["fscore"].max())].reset_index()
    uniques = list(sorted(df[parameter].unique()))#, key=lambda x:int(x.split("-")[0])))

    plt.errorbar([-0.5,-0.3, 0.5], [maxes[maxes[parameter]==uniques[0]]["fscore"].values[0] for _ in range(3)], label="Highest\nMean", color="tab:gray", fmt='--')#, yerr=[0, 0.049717, 0])
    for x, unique in enumerate(uniques[1:]):
        plt.errorbar([0.5+x, 0.7+x, 1.5+x], [maxes[maxes[parameter]==unique]["fscore"].values[0] for _ in range(3)], color="tab:gray", fmt='--')#, yerr=[0, 0.049717, 0])
    
#     display(maxes)
    
    part = df[df["algorithm"].isin(algs)].copy()[["scorefunction","fscore",parameter]]
    sns.boxplot(data=part.iloc[pd.Categorical(part[parameter],list(sorted(part[parameter].unique()))).argsort()], 
                x=parameter, hue="scorefunction", y="fscore", 
                showmeans=True, meanprops={"marker":"s","markerfacecolor":"white", "markeredgecolor":"black","markersize":"5"}, 
                whis=0.9, fliersize=0.5,
               palette=colors)
    
    plt.rcParams['legend.title_fontsize'] = 20
    if "3OFF2" in algs:
        plt.legend(bbox_to_anchor=(1,1), fontsize=15)#, title="Constraint\nAlgorithms")
    else:
        plt.legend(bbox_to_anchor=(1,1), fontsize=15, title="")#Score Metric")
    plt.xlabel(label_name, fontsize=20)
    plt.xticks(fontsize=15)#, rotation=45)
    plt.ylabel("F-Score", fontsize=20)
    plt.yticks(fontsize=15)
#     plt.savefig(f"C:/Users/jacco/Documents/GitHub/Thesis/Figs/Results/Method1/{save_name}/{algs[0]}_{save_name}_accuracy.png", dpi=400, bbox_inches="tight")
    plt.show()
```

# Plot Non-significant Properties

```python
colors = {"HC":"tab:purple", "TABU":"tab:blue", "MMHC":"tab:green", "MIIC":"tab:brown", "3OFF2":"tab:olive"}

label_names = ["Nodes", "Timesteps", "Cardinality"]
for c, col in enumerate(["nodes", "timesteps", "values"]):
    grouped = df[["algorithm","scorefunction","fscore",col]].groupby(["algorithm","scorefunction", col]).mean()[["fscore"]]
    maxes = grouped[grouped["fscore"].isin(grouped.groupby(col)["fscore"].max())].reset_index()
    uniques = list(sorted(df[col].unique()))
    plt.errorbar([-0.5,-0.3, 0.5], [maxes[maxes[col]==uniques[0]]["fscore"].values[0] for _ in range(3)], label="Highest\nMean", color="tab:gray", fmt='--')#, yerr=[0, 0.049717, 0])
    for x, unique in enumerate(uniques[1:]):
        plt.errorbar([0.5+x, 0.7+x, 1.5+x], [maxes[maxes[col]==unique]["fscore"].values[0] for _ in range(3)], color="tab:gray", fmt='--')#, yerr=[0, 0.049717, 0])
    
    part = df[df["scorefunction"].isin(["BDEU","3OFF2","MIIC"])].drop(["scorefunction"],axis=1)
    sns.boxplot(data=part.iloc[pd.Categorical(part["algorithm"],["HC","TABU","MMHC","3OFF2","MIIC"]).argsort()], x=col, hue="algorithm", y="fscore", 
                showmeans=True, meanprops={"marker":"s","markerfacecolor":"white", "markeredgecolor":"black","markersize":"5"},
                palette=colors, whis=0.8, fliersize=0.5)
    plt.legend(bbox_to_anchor=(1,1), fontsize=15, title="")
    plt.xlabel(label_names[c], fontsize=20)
    plt.xticks(fontsize=15)
    plt.ylabel("F-Score", fontsize=20)
    plt.yticks(fontsize=15)
#     plt.savefig(f"C:/Users/jacco/Documents/GitHub/Thesis/Figs/Results/Method1/nodes-timesteps-values/{col}_accuracy.png", dpi=400, bbox_inches="tight")
    plt.show()
```
