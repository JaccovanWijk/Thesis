# Imports

```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 250)
```

# Read Comparison Classifiers

```python
df = pd.read_csv("newOutputs/comparison.csv", index_col=0)
df
```

# Bin Continuous Properties

```python
bins = 4

for col in ["maxdegree/nodes","roots/nodes","leaves/nodes","longestpath/nodes","mean_constraint","mean_degree","PHI","items/PHI"]:
    if len(df[col].unique()) > bins:
        combinations = np.linspace(min(df[col]),max(df[col]),bins+1)
        df[col] = pd.cut(df[col], combinations, include_lowest=True, labels=[f"{np.round(combinations[x],2)}-{np.round(combinations[x+1],2)}" for x in range(len(combinations)-1)]) 
df
```

# Plot Property

```python
parameter = "mean_constraint"
save_name = "constraint"
label_name = "Mean Node-Constraint"

colors = {"L2L":"tab:purple", "AIC":"tab:blue", "BIC":"tab:green", "BDEU":"tab:orange", "K2":"tab:red", "MIIC":"tab:brown", "3OFF2":"tab:olive", "Constraint":"tab:brown"}

all_algorithms = [["HC"], ["TABU"], ["MMHC"],["MIIC","3OFF2"]]
for i, algs in enumerate(all_algorithms):   
#     display(df[["algorithm","scorefunction","difference",parameter]].groupby(["algorithm","scorefunction", parameter]).var()[["difference"]].dropna())
    grouped = df[["algorithm","scorefunction","difference",parameter]].groupby(["algorithm","scorefunction", parameter]).mean()[["difference"]]
    maxes = grouped[grouped["difference"].isin(grouped.groupby(parameter)["difference"].max())].reset_index()
    uniques = list(sorted(df[parameter].unique()))
    plt.errorbar([-0.5,-0.3, 0.5], [maxes[maxes[parameter]==uniques[0]]["difference"].values[0] for _ in range(3)], label="Highest\nMean", color="tab:gray", fmt='--')#, yerr=[0, 0.049717, 0])
    for x, unique in enumerate(uniques[1:]):
        plt.errorbar([0.5+x, 0.7+x, 1.5+x], [maxes[maxes[parameter]==unique]["difference"].values[0] for _ in range(3)], color="tab:gray", fmt='--')#, yerr=[0, 0.049717, 0])
    
    part = df[df["algorithm"].isin(algs)].copy()[["scorefunction","difference",parameter]]
    sns.boxplot(data=part.iloc[pd.Categorical(part[parameter],list(sorted(part[parameter].unique()))).argsort()], 
                x=parameter, hue="scorefunction", y="difference", 
                showmeans=True, meanprops={"marker":"s","markerfacecolor":"white", "markeredgecolor":"black","markersize":"5"}, 
                whis=0.9, fliersize=0.5,
               palette=colors)
    
    plt.rcParams['legend.title_fontsize'] = 20
    if "3OFF2" in algs:
        plt.legend(bbox_to_anchor=(1,1), fontsize=15)#, title="Constraint\nAlgorithms")
    else:
        plt.legend(bbox_to_anchor=(1,1), fontsize=15, title="")#Score Metric")
    plt.xlabel(label_name, fontsize=20)
    plt.xticks(fontsize=15)#, rotation=45)
    plt.ylabel("Difference in F-Score", fontsize=20)
    plt.yticks(fontsize=15)
#     plt.savefig(f"C:/Users/jacco/Documents/GitHub/Thesis/Figs/Results/Method2/{save_name}/{algs[0]}_{save_name}_accuracy.png", dpi=400, bbox_inches="tight")
    plt.show()
```
